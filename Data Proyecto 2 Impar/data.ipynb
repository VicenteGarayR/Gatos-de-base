```python
pago_compra = df_pagos_v2[df_pagos_v2['subs_id'].isna()]
pago_compra = pago_compra.drop(columns=['subs_id'])
pago_compra["pid"] = pago_compra["pid"].astype(int)
pago_compra["pro_id"] = pago_compra["pro_id"].astype(int)
pago_compra
```

```python
tabla_visualizaciones_capitulos = vis_series[['uid', 'cid', 'fecha']]
tabla_visualizaciones_capitulos["cid"] = tabla_visualizaciones_capitulos["cid"].astype(int)
tabla_visualizaciones_capitulos["fecha"] = pd.to_datetime(tabla_visualizaciones_capitulos["fecha"])
tabla_visualizaciones_capitulos #PK: No hay porque es una relacion entre capitulos y usuarios, por lo tanto pueden haber capitulos con el mismo usuario y viceversa.
```

```python
prov_pelis1 = prov_pelis[['id', 'nombre', 'costo']].rename({'nombre':'proveedor', 'id': 'pro_id'}, axis=1)
prov_series1 = prov_series[['id', 'nombre', 'costo']].rename({'nombre':'proveedor', 'id': 'pro_id'}, axis=1)
prov_pelis_pextra = prov_pelis_pextra[['id', 'nombre', 'costo']].rename({'nombre':'proveedor','id': 'pro_id'}, axis=1)
subs_cancel1 = subs_canceladas[['pro_id', 'proveedor', 'costo']]
subs_activas1 = subs_activas[['pro_id', 'proveedor', 'costo']]
```

```python
tabla_proveedor_costos = pd.merge(prov_pelis1, prov_series1, how='outer', on=['pro_id', 'proveedor', 'costo'])
tabla_proveedor_costos = pd.merge(tabla_proveedor_costos, prov_pelis_pextra, how='outer', on=['pro_id', 'proveedor', 'costo'])
tabla_proveedor_costos = pd.merge(tabla_proveedor_costos, subs_cancel1, how='outer', on=['pro_id', 'proveedor', 'costo'])
tabla_proveedor_costos = pd.merge(tabla_proveedor_costos, subs_activas1, how='outer', on=['pro_id', 'proveedor', 'costo'])
tabla_proveedor_costos = tabla_proveedor_costos.drop_duplicates(subset= "pro_id", keep='first', ignore_index=True).sort_values(by='pro_id')
tabla_proveedor_costos
```
